```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_sensitivity_analysis(df, title, curve_label, x_axis_label, y_axis_label, x_col, y_col, base_case_val):
    fig, ax = plt.subplots(figsize=(10,6))
    ax.set(title=title, xlabel=x_axis_label, ylabel=y_axis_label)
    plt.plot(df[x_col], df[y_col], color='b',label=curve_label)
    plt.axhline(y=50000, color='orange', linestyle='-', label='Willingness-to-Pay Threshold')
    plt.axvline(x=base_case_val, color='red', linestyle='dotted', label='Base Case Value')
    plt.legend()
    plt.show()
```

```python
p_oto = pd.read_csv("base_case/pOtotoxicity2.csv")
p_oto.head()
```

```python
title = "ICER of Enhanced Clinical Observation compared to NSC vs. Probability of Ototoxicity"
curve_label = "ICER of Enhanced Clinical Observation compared to NSC"
x_axis_label = "Probability of Ototoxicity"
y_axis_label = "ICER ($/QALY)"
x_col = "Probability of Ototoxicity"
y_col = "ICER of Enhanced Clinical Observation compared to NSC"

plot_sensitivity_analysis(p_oto, title, curve_label, x_axis_label, y_axis_label, x_col, y_col, 0.014)
```
